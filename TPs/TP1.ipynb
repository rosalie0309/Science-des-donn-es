### D- Génération des données et affichage 

```python
# 1- utilisons help pour voir comment utiliser datasets.make_blobs
help(datasets.make_blobs)
```

```python
# 2- Générez 1000 données de deux variables réparties en 4 groupes. 

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, n_features=2, centers=4)
```

```python
# Utilisez les fonctions figure, scatter, title, xlim, ylim, xlabel, ylabel et show pour afficher 
# les données avec des couleurs correspondant aux classes. Les axes x et y seront dans 
# l’intervalle [-15, 15] et devrons avoir un titre. La figure doit aussi avoir un titre.

plt.figure(figsize=(7,4))
plt.scatter(X[:,0], X[:,1], c=y, cmap="viridis", marker='o')
plt.title("Visualisation des données générées avec make_blobs", loc='center', c="orange")
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.show()
```

```python
#Générez 100 données de deux variables réparties en 2 groupes, puis 500 données de deux 
#variables réparties en 3 groupes. Concaténez (vstack et hstack) les deux jeux de données et 
#les numéros de classe pour fusionner les deux jeux de données. Affichez les trois 
#ensembles avec scatter comme précédemment. 

# Génération des deux jeux de données 
X1, y1 = make_blobs(n_samples=100, n_features=2, centers=2)

X2, y2 = make_blobs(n_samples=500, n_features=2, centers=3)

# Concatenation des deux jeux de données 
# Appliquons un empilement vertical pour les données 
X1_X2_vertical = np.vstack((X1, X2)) 

# Concatenation des numéros de classe: on applique ici un empilement horizontal
# Nous allons décaler les labels du deuxième jeu de données de 1 avant de concatener afin de bien voir la différence dans le troisième
# jeu de données 
y2_decale = y2 + y1.max()+1
y1_y2_horizontal = np.hstack((y1,y2_decale))

# Affichons les 3 jeux de données 

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.scatter(X1[:,0], X1[:,1], c=y1, marker='o', cmap='viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.xlabel("variable 1")
plt.ylabel("variable 2")
plt.title("premier jeu de données", loc="center", c="orange")

plt.subplot(1,3,2)
plt.scatter(X2[:,0], X2[:,1], c=y2_decale, marker='o', cmap='viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.xlabel("variable 1")
plt.ylabel("variable 2")
plt.title("deuxième jeu de données", loc="center", c="orange")

plt.subplot(1,3,3)
plt.scatter(X1_X2_vertical[:,0], X1_X2_vertical[:,1], c=y1_y2_horizontal, marker='o', cmap='viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.xlabel("variable 1")
plt.ylabel("variable 2")
plt.title("troisième jeu de données", loc="center", c="orange")
plt.show()

```
